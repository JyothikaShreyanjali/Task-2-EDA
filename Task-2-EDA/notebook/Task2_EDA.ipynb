```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df = df.drop(columns=["Cabin"], errors="ignore")
```

```python
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df = pd.get_dummies(df, columns=["Embarked"], drop_first=True)
```

```python
df = df.drop(columns=["Name", "Ticket", "PassengerId"], errors="ignore")
```

```python
df.head()
df.info()
df.isnull().sum()
```

```python
df.to_csv("cleaned_titanic.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing survival counts
sns.countplot(x='Survived', data=df)
plt.title('Distribution of Survival')
plt.show()

# Get exact percentages
print(df['Survived'].value_counts(normalize=True))
```

```python
plt.figure(figsize=(10, 6))
# df.corr() measures linear relationships between columns
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
# Overlaying age distributions for survivors vs non-survivors
sns.histplot(data=df, x='Age', hue='Survived', kde=True, multiple='stack')
plt.title('Age Distribution by Survival Status')
plt.show()
```
